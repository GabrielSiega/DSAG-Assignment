### Given a specific series of values 1, 1, 2, 6, 24, 120, and so on, your task is to write an efficient program that generates this sequence. Please provide code in the code cell below to generate the sequence and an expected output is as shown below:

### Your code should be written in a single cell below

```python
def generate_sequence(length):
    sequence = [1]  # Start with the initial value of 1

    for i in range(1, length):
        # Multiply the previous term by the next consecutive positive integer
        if i == 1:
            sequence.append(sequence[-1])
        else:
            sequence.append(sequence[-1] * i)

    return sequence

# Specifying desired length of the sequence
sequence_length = 31  # You can change this to any positive integer

# Generating the sequence
result_sequence = generate_sequence(sequence_length)

# Print the generated sequence with term numbers from top to bottom
for i, value in enumerate(result_sequence):
    print(f'term {i}: {value}')
```

```python
def clean_string(input_str):
    cleaned_chars = [char.lower() for char in input_str if char.isalpha()]
    return ''.join(cleaned_chars)

def is_palindrome(input_str, by_word_unit=False):
    clean_str = clean_string(input_str)
    if by_word_unit:
        words = clean_str.split()
        return words == words[::-1]
    return clean_str == clean_str[::-1]

def print_palindrome_menu():
    options = ["Palindrome", "Palindrome by word unit"]
    counter = 1
    
    print()
    print("Fun with Palindromes Menu")
    print()

    for option in options:
        print(f"{counter}. {option}")
        counter += 1

while True:
    # Display menu
    print_palindrome_menu()

    # Get user input
    user_choice = input("Enter 1 or 2 (or 'q' to quit): ")
    print()
    print("You chose: ",user_choice)

    # Check user choice and perform corresponding action
    if user_choice in ["1", "2"]:
        user_input = input("Enter a string: ")
        print("Your String Input: ",user_input)
        result = "palindrome" if is_palindrome(user_input, user_choice == "2")else "not a palindrome"
        print()
        print(f"The input is a {result}.")
    elif user_choice.lower() == 'q':
        print("Quitting the program.")
        break
    else:
        print("Invalid choice. Please enter 1 or 2 (or 'q' to quit).")
```

### Develop a quiz generation software program using only the 'random' module. You're provided with a text file containing 50 quiz questions, which are of two types: Multiple-Choice Questions (MCQ) and True/False questions. Some MCQs may have varying numbers of answer options. The correct answers are indicated in the file. Access the quiz text file through the given link [here](https://studenttpedu-my.sharepoint.com/:t:/g/personal/kokyau_tp_edu_sg/EaYKgwHVtAZJhCLfCjyeTNQBxNlk4azYAEJMLMunk1Pmzw?e=Xn5c26)

### When the program runs, it will prompt users for specific parameters, including the number of questions to generate, whether questions should be randomized, and whether answer options should be randomized. When users answer questions incorrectly, the program will display the correct answers. At the end of the quiz, the total score will be presented, and users will have the option to retake the quiz.

### You must strictly use the 'random' module, and no other modules are allowed. The provided text file must not be modified, as a separate text file with different questions will be used for grading. For a clear understanding of how your software should work, please view the video demonstration at the link [here](https://studenttpedu-my.sharepoint.com/:v:/g/personal/kokyau_tp_edu_sg/ET8e8-4fXBhNlw6B7p8qB9UB5gyfTryYE8jkKUtvOuDUjw?e=fWKJWU) The video illustrates the expected behavior of your program.
### Your code should be written in a single cell below

```python
import random
import os

file_path = os.path.join(os.getcwd(), r"DSAG-Assignment\quiz_questions.txt")

def load_questions(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    questions = []
    current_question = None

    for line in lines:
        line = line.strip()
        if line.startswith("Q:"):
            if current_question:
                questions.append(current_question)
            current_question = {'question': line[3:], 'options': [], 'answer': ''}
        elif line.startswith("A:"):
            current_question['answer'] = line[3:]
        elif line.startswith("MCQ:"):
            options = line[5:].split('|')
            current_question['options'] = options
        elif line.startswith("TF:"):
            current_question['options'] = ['True', 'False']

    if current_question:
        questions.append(current_question)

    return questions

def generate_quiz(questions, num_questions, randomize_questions, randomize_options):
    if randomize_questions:
        questions = random.sample(questions, num_questions)
    else:
        questions = questions[:num_questions]

    score = 0

    for question in questions:
        print(f"{question['question']}")
        if randomize_options:
            options = random.sample(question['options'], len(question['options']))
        else:
            options = question['options']

        for option in options:
            print(f"   {option}")

        user_answer = input("Enter the letter corresponding to your answer: ").lower()

        if user_answer.isalpha() and user_answer.islower() and 'a' <= user_answer <= 'z':
            if user_answer.strip() == question['answer'].lower().strip():
                print("Correct!\n")
                score += 1
            else:
                print(f"Incorrect. The correct answer is: {question['answer']}\n")
        else:
            print("Invalid input. Please enter a valid letter.")

    return score

def main():
    file_path = "quiz_questions.txt"  # Update with the correct file path
    questions = load_questions(file_path)

    while True:
        try:
            num_questions = int(input("Enter the number of questions to generate: "))
            randomize_questions = input("Randomize questions? (yes/no): ").lower() == 'yes'
            randomize_options = input("Randomize answer options? (yes/no): ").lower() == 'yes'
        except ValueError:
            print("Invalid input. Please enter a valid number.")
            continue

        score = generate_quiz(questions, num_questions, randomize_questions, randomize_options)

        print(f"\nYour total score: {score}/{num_questions}\n")

        retake = input("Do you want to retake the quiz? (yes/no): ").lower()
        if retake != 'yes':
            break

if __name__ == "__main__":
    main()
```
